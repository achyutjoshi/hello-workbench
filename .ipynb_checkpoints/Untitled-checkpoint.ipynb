```python
import string

def clean_string(in_string):
    table = str.maketrans('', '', string.punctuation)
    clean_string = in_string.translate(table)
    clean_string = ''.join([i for i in clean_string if not i.isdigit()])
    clean_string = clean_string.strip() #remove leading and trailing spaces
    return(clean_string)
```
